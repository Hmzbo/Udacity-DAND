## Dataset Overview

In this dataset exploration project, we analyzed the data of a bike sharing system located in the greater San Francisco Bay area which cover three regions San Francisco, East Bay, and San Jose. In the present dataset we have informations are about 4,890,404 trips taken from the year 2017 to 2019.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import os
from math import sin, cos, sqrt, atan2, radians

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
all_files = os.listdir()
csv_files = [item for item in all_files if item[len(item)-4:len(item)]=='.csv' ]
li=[]
for filename in csv_files:
    temp_df = pd.read_csv(filename, index_col=None)
    temp_df['year']=str(filename[0:4])
    li.append(temp_df)

df = pd.concat(li, axis=0, ignore_index=True)
drop_list=['rental_access_method','bike_share_for_all_trip','start_station_latitude','start_station_longitude','end_station_latitude','end_station_longitude','bike_id']
df.drop(drop_list, axis=1, inplace=True)
max_dur = df.duration_sec.max()
df.query('duration_sec==@max_dur')
idx = df[df['start_station_id'].isnull()].index
df.drop(idx, inplace=True)
```

## (Visualization 1)
### How long does the average trip takes?

```python
df['duration_sec'].mean()/60
```

Generally, the average duration of the trips taken in this sharing bike system is: 14.27 minutes.
However, further investigation of the dataset revealed the existance of two trip duration categories: Medium and Long trips, the distribution of the trip duration in seconds is shown in the figure below.

```python
fig = plt.figure(figsize=(15,6))

p25=np.percentile(df["duration_sec"],25)
p75=np.percentile(df["duration_sec"],75)
IQR=p75-p25
upper_whisker = df[df["duration_sec"]<=p75+1.5*IQR]["duration_sec"].max()

df_temp_long=df.query('@upper_whisker < duration_sec')
df_temp_medium=df.query('duration_sec <= @upper_whisker')

max_log_m = np.log10(df_temp_medium['duration_sec']).max()
min_log_m = np.log10(df_temp_medium['duration_sec']).min()
max_log_l = np.log10(df_temp_long['duration_sec']).max()
min_log_l = np.log10(df_temp_long['duration_sec']).min()

binss_long = 10 ** np.arange(min_log_l,max_log_l+0.1,0.1)
binss_medium = 10 ** np.arange(min_log_m,max_log_m+0.1,0.1)

ticks_m = list(np.arange(10**min_log_m,10**max_log_m,500))
label_m=['{}'.format(int(v)) for v in ticks_m]

ticks_l = list(np.arange(10**min_log_l,10**max_log_l,3*10**4))
label_l=['{}'.format(int(v)) for v in ticks_l]

fig.suptitle('Historam of duration classes (Medium Vs Long)', fontsize=15)
ax1 = plt.subplot(121)
plt.hist(df_temp_medium['duration_sec'],bins=binss_medium)
ax1.set_title('Histogram of medium duration class trips')
plt.xscale('log')
plt.xticks(ticks_m,label_m)
plt.xlabel('Trip duration (sec)')
plt.ylabel('frequency')
ax2=plt.subplot(122)
plt.hist(df_temp_long['duration_sec'],bins=binss_long)
ax2.set_title('Histogram of long duration class trips')
plt.xscale('log')
plt.xticks(ticks_l,label_l)
plt.xlabel('Trip duration (sec)')
plt.ylabel('frequency');
```

```python
df['duration_class']='Medium'
idx_m = df.query('duration_sec <= @upper_whisker').index
idx_l = df.query('@upper_whisker < duration_sec').index
df.loc[idx_m,'duration_class']='Medium'
df.loc[idx_l,'duration_class']='Long'
```

Treating each duration class apart would give us a better idea about the average duration of the trips redcorded in this dataset. The figures below show the proportion of each class in our data set, and the average duration for each class in seconds.

```python
dc = df['duration_class'].value_counts()
plt.figure(figsize=(14,5))
plt.subplot(121)
plt.pie(x=dc, labels=dc.index, autopct='%1.1f%%',shadow=True, startangle=90)
plt.title('duration_class')
plt.subplot(122)
sb.barplot(data=df, x='duration_class',y='duration_sec', color=sb.color_palette()[0])
plt.title('Average duration for each duration class');
```

## (Visualization 2)
### When are most trips taken in terms of day of the week?

```python
week_days={
    '0': 'Monday',
    '1': 'Tuesday',
    '2': 'Wednesday',
    '3': 'Thursday',
    '4': 'Friday',
    '5': 'Saturday',
    '6': 'Sunday'
}

df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])
df['trip_day'] = df.start_time.dt.dayofweek
for i in list(week_days.keys()):
    df.trip_day.replace(int(i),week_days[i],inplace=True)
dotw_s = df['trip_day'].value_counts()
```

```python
l=[week_days[item] for item in week_days.keys()]
dotw=pd.api.types.CategoricalDtype(categories=l, ordered=True)
df['trip_day']=df['trip_day'].astype(dotw)
```

Getting insights about the trips in each day of the week would help the Ford GoBike Comapny to take effective measure to facilitate the flow of the system.

The first plot in the figure below shows the number of trips taken during each day of the week in the last three years. It is clear that the number of trips taken during the weekend days is on average about half the number of trips taken during workdays.

Further investigation, showed that the average duration of trips taken during the weekends is about 50% longer than the average  duration of trips taken in the rest of the week.

```python
fig = plt.figure(figsize=(15,5))
fig.suptitle('Number of trips Vs Average duration', fontsize=15)
plt.subplot(121)
sb.barplot(data = dotw_s.reset_index(), x='index', y='trip_day', palette='Blues')
plt.xlabel('Days of the week')
plt.ylabel('Number of trips')
plt.xticks(rotation=45)
plt.subplot(122)
sb.barplot(data=df, x='trip_day',y='duration_sec', palette='Blues' )
plt.xlabel('Days of the week')
plt.ylabel('Trip duration on average (sec)')
plt.xticks(rotation=45);
```

**Insight:** The users of the Ford GoBike System would more likely go for a long bike trip during the weekends to hangout or as some kind of workout exercice. While during the workdays, the shorter average trip duration could be explained by the needs of users for short rides, like going to work or going to groceries..etc.


The figure below shows the that the observation made above about the number of trips and the average duration is maintained during the three years.

```python
fig = plt.figure(figsize=(15,6))
fig.suptitle('Evolution of the average duration and count of trips for each day by year', fontsize=15)
plt.subplot(121)
sb.barplot(data=df, x='year', y='duration_sec', hue='trip_day')
plt.xlabel('Year')
plt.ylabel('Average duration (sec)')
dotw_trips_per_year = df.groupby('year')['trip_day'].value_counts()
df_temp = dotw_trips_per_year.to_frame()
df_temp.rename(columns={'trip_day':'nbr_trips'},inplace=True)
df_temp = df_temp.reset_index()
plt.subplot(122)
sb.barplot(data=df_temp, x='year', y='nbr_trips', hue='trip_day')
plt.xlabel('Year')
plt.ylabel('Number of trips');
```

## (Visualization 3)
### What are the most frequented stations?

Knowing the most frequented stations would help the company take measures to make system more efficient.

The figures below shows the top 5 most frequented start and end stations in our dataset.

```python
df['station_freq']='Noraml'
hf_end_st_id = list(df['end_station_id'].value_counts()[df['end_station_id'].value_counts()>33000].index)
hf_start_st_id = list(df['start_station_id'].value_counts()[df['start_station_id'].value_counts()>33000].index)
hf_start_end_st_id = list(set(hf_end_st_id)&set(hf_start_st_id))

idx1 = df.query('end_station_id in @hf_end_st_id').index
idx2 = df.query('end_station_id in @hf_start_st_id').index
idx3 = df.query('end_station_id in @hf_start_end_st_id').index

df.loc[idx1,'station_freq']='High_end'
df.loc[idx2,'station_freq']='High_start'
df.loc[idx3,'station_freq']='High_start_end'

top5_end = hf_end_st_id[-5:]
top5_start = hf_start_st_id[-5:]
hf_end_st_names = df.query('end_station_id in @top5_end')['end_station_name'].unique()
hf_start_st_names = df.query('start_station_id in @top5_start')['start_station_name'].unique()

hf_end_st = df['end_station_id'].value_counts().sort_values().tail(5)
hf_start_st = df['start_station_id'].value_counts().sort_values().tail(5)



plt.figure(figsize=(15,5))
plt.subplot(121)
ax2 = hf_start_st.plot(kind='bar', color=sb.color_palette('Greens')[0:5])
ax2.set_xticklabels(list(hf_start_st_names))
plt.xlabel("Station name")
plt.title('Top 5 highly frequented start station')
plt.ylabel("Frequency")
plt.xticks(rotation=65)
plt.subplot(122)
ax = hf_end_st.plot(kind='bar', color=sb.color_palette('Reds')[0:5])
ax.set_xticklabels(list(hf_end_st_names))
plt.xlabel("Station name")
plt.ylabel("Frequency")
plt.xticks(rotation=65)
plt.title('Top 5 highly frequented end station');
```

**Insights:**
- The most frequented start station is Bancroft Way at College Ave with more than 85,000 trips starting there during three years. This station is one of four stations that are the closest to the University of California Berkley and many other institutions. 
- In the other hand, the most frequented end station is Folsom St at 3rd St with almost 120,000 trips ending there during three years. This station is in a residential area which suggest that a good number users of this bike sharing system lives near this area, or visit it frequently for any other reason.

It's important to note that more than 3 million trip during these three years are made from and to a none highly frequenty stations which means stations with less than 33,000 trips during this period.

## (Visualization 4)

### Does the duration of the trips  depends on the year or user type?

Revealing the hidden informations in this dataset about the relation about the trip duration and user type would the Ford GoBike company make the right decision to increase their revenue.

The charts below shows the difference between both types of users in term on their count and the average duration of their trips, and the evolution of the number of customers and subscribers each year.

```python
users = df['user_type'].value_counts()
fig = plt.figure(figsize=(15,4))
fig.suptitle('Count and Average duration by user type', fontsize=15)
plt.subplot(121)
users.plot(kind='bar',rot=0, color=sb.color_palette()[0])
plt.xlabel('User type')
plt.ylabel('Count')
plt.subplot(122)
sb.barplot(data=df, x='user_type',y='duration_sec', color=sb.color_palette()[0])
plt.xlabel('User type')
plt.ylabel('Average duration (sec)');
```

```python
plt.figure(figsize=(10,6))
plt.title('Evolution of the number of users by year for each user type', fontsize=15)
sb.countplot(data=df, x='year',hue='user_type', palette=sb.color_palette())
plt.xlabel('Year')
plt.ylabel('Count');
```

The number of trips made by both type of users increased each year with an immense increase in the trips made by subscribers from 2017 to 2018, which implies that this system gained fame and reputation during these years.

Moreover, the total number of trips made by subscribers is more than 4 times the number of trips made by customers, whereas the average duration of the trips made by customers is slightly more than two times the average duration made by subscribers.

**Insights:** This comparaison suggest that the subscribers are more active in the system and take trips more often but for short duration; For example, going to work, to the gym, or to do groceries..etc. While the customers tends to go for trip less often but for longer duration; For example, going for a long ride during weekends, taking a trip to hangout, or simple exercising..etc.
