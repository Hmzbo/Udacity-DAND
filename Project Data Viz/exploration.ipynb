## Preliminary Wrangling

The Ford GoBike System data set includes informations about individual rides made in a bike-sharing system covering the greater San Francisco Bay area from 2017 to 2019.


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import os
from math import sin, cos, sqrt, atan2, radians

%matplotlib inline
```

```python
all_files = os.listdir()
csv_files = [item for item in all_files if item[len(item)-4:len(item)]=='.csv' ]
```

```python
li=[]
for filename in csv_files:
    temp_df = pd.read_csv(filename, index_col=None)
    temp_df['year']=str(filename[0:4])
    li.append(temp_df)

df = pd.concat(li, axis=0, ignore_index=True)
```

```python
df.info()
```

```python
drop_list=['rental_access_method','bike_share_for_all_trip','start_station_latitude','start_station_longitude','end_station_latitude','end_station_longitude','bike_id']
df.drop(drop_list, axis=1, inplace=True)
```

### What is the structure of your dataset?

Our data set contains informations about 4890404 trips in the Ford GoBike System. The informations we have can be categorized in 3 categories:
   - Informations about the time: 
        - duration_sec, start_time, end_time, year
   - Informations about the location:
        - start_station_id, start_station_name, start_station_latitude, start_station_longitude, end_station_id, end_station_name, end_station_latitude, end_station_longitude
   - Informations about user:
        - bike_id, user_type

### What is/are the main feature(s) of interest in your dataset?

The main feature of interest in this dataset is to have a better understanding of the workflow of the Ford GoBike System which can be achieved by analyzing the data to answer questions like:
- How long does the average trip takes?
- When are most trips taken in terms of day of the week?
- What are the most frequented stations?
- Does the duration of the trips  depends on the year or user_type?


### What features in the dataset do you think will help support your investigation into your feature(s) of interest?

Almost every variable in this dataset will help us support our investigation. We are only excluding the longitude/latitude, and the bike_id variables since they do not support the analysis we're going to do.

As shown in the plots above, the distribution variable duration_sec which describes the duration of a trip in seconds is extremely right skewed which implies the existence of many outliers. Furher investigation of the trip corresponding to the maximum duration shows that this data lacks informations about the start and end stations. In order make sure that the data we're analysing is reliable, we will use only the data about trips that we have all the necessary informations about.

Moreover, Given that the duration_sec variable has a lot of outliers, we're going to create a new categorical variable that labels the trips with duration less than the `upper_quartile + 1.5 * IQR`  as Medium, and labels the trips with duration more than the `upper_quartile + 1.5 * IQR`  as Long. That way we can analyse each category individually.

```python
fig = plt.figure(figsize=(15,10))


p25=np.percentile(df["duration_sec"],25)
p75=np.percentile(df["duration_sec"],75)
IQR=p75-p25
upper_whisker = df[df["duration_sec"]<=p75+1.5*IQR]["duration_sec"].max()

df_temp_long=df.query('@upper_whisker < duration_sec')
df_temp_medium=df.query('duration_sec <= @upper_whisker')
binss_long=np.arange(upper_whisker,max_dur+35,35)
binss_medium=np.arange(0,upper_whisker+35,35)

fig.suptitle('Boxplot/Historam for duration_sec variable (Medium Vs Long, duration)', fontsize=15)
ax1 = plt.subplot(221)
plt.boxplot(df_temp_medium['duration_sec']);
ax1.set_title('Boxplot of medium trips')
ax2=plt.subplot(222)
plt.hist(df_temp_medium['duration_sec'],bins=binss_medium);
ax2.set_title('Histogram of medium trips')
ax3=plt.subplot(223)
plt.boxplot(df_temp_long['duration_sec']);
ax3.set_title('Boxplot of long trips')
ax4=plt.subplot(224)
plt.hist(df_temp_long['duration_sec'],bins=binss_long)
ax4.set_title('Histogram of long trips');
```

The right skewness of the distribution is maintained even after the creation of two categories, however the number of outliers is severly reduced for the Medium duration category which make it possible for us to extract better information about this category on average. In the other hand, the distribution of the Long duration category inherited the extreme righ skewness characteristic. We will keep the latter category on its current state, since we are more interested in the Medium duration category wich represent more than 75% of the data in hand.

```python
df['duration_class']='Medium'
idx_m = df.query('duration_sec <= @upper_whisker').index
idx_l = df.query('@upper_whisker < duration_sec').index
df.loc[idx_m,'duration_class']='Medium'
df.loc[idx_l,'duration_class']='Long'
```

As a second exploration, we targeted the variables start_station_id, and start_station_id individually. We used the bar plot to see the frequency of each staion. Since we have more than 400 stations in our dataset we created two sub-groups after ordering them by ascending frequency, the separation threshhold we used is the frequency median. We can clearly observe in the second sub-plot of in each figure that there is a change in the increase rate of the frequency starting from the frequency 33000 shown by the green line. Our investigation will focus more on the stations with frequency greater than 33000. Thus we need to create a new categorical variable named `station_freq` which indicates whether a certain trip was tanken between highly frequented stations or no.

```python
hf_end_st = df['end_station_id'].value_counts()[df['end_station_id'].value_counts()>33000].sort_values()
hf_start_st = df['start_station_id'].value_counts()[df['start_station_id'].value_counts()>33000].sort_values()
plt.figure(figsize=(14,5))
plt.subplot(121)
hf_end_st.plot(kind='bar')
plt.xlabel("station id")
plt.ylabel("frequency")
plt.title('Highly frequented end station')
plt.subplot(122)
hf_start_st.plot(kind='bar')
plt.xlabel("station id")
plt.title('Highly frequented start station')
plt.ylabel("frequency");
```

```python
df['station_freq']='Noraml'
hf_end_st_id = list(df['end_station_id'].value_counts()[df['end_station_id'].value_counts()>33000].index)
hf_start_st_id = list(df['start_station_id'].value_counts()[df['start_station_id'].value_counts()>33000].index)
hf_start_end_st_id = list(set(hf_end_st_id)&set(hf_start_st_id))

idx1 = df.query('end_station_id in @hf_end_st_id').index
idx2 = df.query('end_station_id in @hf_start_st_id').index
idx3 = df.query('end_station_id in @hf_start_end_st_id').index

df.loc[idx1,'station_freq']='High_end'
df.loc[idx2,'station_freq']='High_start'
df.loc[idx3,'station_freq']='High_start_end'
```

```python
top5_end = hf_end_st_id[-5:]
top5_start = hf_start_st_id[-5:]
hf_end_st_names = df.query('end_station_id in @top5_end')['end_station_name'].unique()
hf_start_st_names = df.query('start_station_id in @top5_start')['start_station_name'].unique()

hf_end_st = df['end_station_id'].value_counts().sort_values().tail(5)
hf_start_st = df['start_station_id'].value_counts().sort_values().tail(5)

plt.figure(figsize=(8,14))
plt.subplot(211)
ax = hf_end_st.plot(kind='barh')
ax.set_yticklabels(list(hf_end_st_names))
plt.xlabel("station id")
plt.ylabel("frequency")
plt.title('Top 5 highly frequented end station')
plt.subplot(212)
ax2 = hf_start_st.plot(kind='barh' )
ax2.set_yticklabels(list(hf_start_st_names))
plt.xlabel("station id")
plt.title('Top 5 highly frequented start station')
plt.ylabel("frequency");
```
